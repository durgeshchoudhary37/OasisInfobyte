```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
df = pd.read_csv('Advertising.csv')
df.head()
```

```python
df.columns
```

```python
X = df[['TV', 'Radio', 'Newspaper']]
y = df['Sales']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size =0.3, random_state=42)
gb = GradientBoostingRegressor(n_estimators=100, random_state=42)
gb.fit(X_train, y_train)
```

```python
y_hat = gb.predict(X_test)
```

```python
mse = mean_squared_error(y_test, y_hat)
mae = mean_absolute_error(y_test, y_hat)
print("Gradient Boosting Mean Squared Error:", mse)
print("Gradient Boosting Mean Absolute Error:", mae)
```

```python
plt.scatter(y_test, y_hat)
plt.xlabel('Actual Sales')
plt.ylabel('Predicted Sales (Gradient Boosting)')
plt.title('Actual vs Predicted Sales (Gradient Boosting)')
plt.show()
```

```python
# Visualize the results with custom colors
plt.scatter(y_test, y_hat, color='blue', label='Predicted Sales')  # Predicted sales points in blue
plt.scatter(y_test, y_test, color='red', label='Actual Sales')  # Actual sales points in red
plt.xlabel('Actual Sales')
plt.ylabel('Predicted Sales (Gradient Boosting)')
plt.title('Actual vs Predicted Sales (Gradient Boosting)')
plt.legend()
plt.show()
```
